#### Question 3: 
The following functions are all supposed to count how many times a certain base (represented as a character variable in Python) appears in a dna sequence (represented as a string variable in Python): 

Which of them is correct?

```python
def count1(dna, base): # appears to work well 
    i = 0
    for c in dna:
        if c == base:
    	    i += 1 
    return i

def count2(dna, base): #works well 
    i = 0 
    for j in range(len(dna)): #creates an index of 0-9 
        if dna[j] == base:
    	    i += 1 
    return i 

def count3(dna, base): #works well! 
    match = [c == base for c in dna] #list comprehension generates a true/false list. sum returns sum of trues. pretty cool! 
    return sum(match)

def count4(dna, base): #works well. very clean. 
    return dna.count(base)

def count5(dna, base): #works 
    return len([i for i in range(len(dna)) if dna[i] == base]) #gives indices of true values 

def count6(dna,base):
    return sum(c == base for c in dna) #same as count 3 
```

```python
base = 'c'

count4(dna, base)
```

# making a repeat finder

(4) A repeat is a substring of a DNA sequence that occurs in multiple copies (more than one) somewhere in the sequence. Although repeats can occur on both the forward and reverse strands of the DNA sequence, we will only consider repeats on the forward strand here. Also we will allow repeats to overlap themselves. For example, the sequence ACACA contains two copies of the sequence ACA - once at position 1 (index 0 in Python), and once at position 3. Given a length n, your program should be able to identify all repeats of length n in all sequences in the FASTA file. Your program should also determine how many times each repeat occurs in the file, and which is the most frequent repeat of a given length.

```python
from itertools import combinations_with_replacement
```

```python
def repeat_finder(name, seq, n):
    repeat_count_dict = {}
    seq = seq.upper()
    combinations = list(itertools.product('ATGC', repeat=n))
    #for combination in combinations:
        #print(combination)
    
    queries = ["".join(combination) for combination in combinations]
    
    repeat_list = []
    #print(queries)
    for query in queries:
        counts = seq.count(query)
        repeat_list.append([name, n, query, counts])
    return repeat_list
```

# remaking a dataframe ORF finder 

```python
import re
my_seq =  'taaatgatgatgtagtatgagtagttataa'

from functools import reduce

def find_substring_indices(string, substrings):
    indices = []
    for substring in substrings:
        indices += ([m.start() for m in re.finditer(substring, string)]) #finds all stop codons and adds them to the stop indices 
    return indices

def reading_frame(number):
    if (number)  % 3 == 0:
        #print ('1st reading frame')
        frame = 1
        #return frame
    elif (number -1) % 3 == 0:
        #print ('2nd reading frame')
        frame = 2
        #return frame
    elif (number -2) % 3 == 0:
        #print ('3rd reading frame')
        frame = 3
    return frame

def orf_finder(name, seq):
    orf_list = []
    
    seq = seq.upper() #convert to uppercase 
    start_indices = [m.start() for m in re.finditer('ATG', seq)] #get indices of all ATG codons, where 0 is the first 
    stop_indices = find_substring_indices(seq, ['TAG','TTA', 'TAA'])#get indices of all stop codons, where 0 is the first 

    #print (start_indices)
    for start in start_indices: #for each start codon 
                
        frame_start = reading_frame(start)

        #check every subsequent codon for a STOP CODON
        i = 0
        while seq[(start+i):(start+i+3)] not in ['TAA', 'TAG', 'TGA']: 
            #print(seq[(start+i):(start+i+3)])
            i +=3
            #print (i, "not a stop")
            if i > len(seq): #will stop if i exceeds the sequence length; i.e. can't find a stop codon 
                break
        else:
            #print ('stop codon found at ',start+i)
            stop = start+i
            orf_seq = seq[start:stop]
            len_orf = len(orf_seq)+3
            orf_list.append([name, start+1, stop+1,frame_start, orf_seq, len_orf])
            

    return orf_list 
        
        
 



# putting it into action 
```
